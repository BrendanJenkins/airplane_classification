# **Photo Pre-Processing Notebook for Aircraft Classifcation Images**

This notebook contains code to crop out copyright tags from our images, to help improve model performance. This was a recommendation from the data's source. 

All that is required to run on your machine is to update cell 2 to contain the proper path to the folder that holds the folder of aircraft images. It will automatically create a new sub-directory called 'cropped' that will have all the pre-processed images. Keep in mind that this will double the storage required on your hard-drive/ssd. 

```python
import os
import numpy as np
import pandas as pd
import torch as nn
from PIL import Image
```

```python
# Remove next two lines if not using google coloab 
from google.colab import drive
drive.mount('/content/gdrive')

# Where the data is stored
data_path = 'gdrive/MyDrive/top_gun_dl/repo/data/'
```

```python
# This function creates a dataframe from the data reposisotry with 
# filepaths and aircraft classification information
def textfile_to_dataframe(path, filename, dataset):
  with open(data_path + filename) as f:
    data_lines = f.readlines()
  
  data = []
  for line in data_lines:
    line_list = line.split(maxsplit=1)
    img_num = line_list[0] + '.jpg'
    label = line_list[1]
    label = label.strip('\n')
    image_path = 'gdrive/MyDrive/top_gun_dl/repo/data/images/' + img_num
    row = [img_num, dataset, label, image_path]
    data.append(row)
  
  df = pd.DataFrame(data, columns=['image', 'dataset', 'label', 'path'])
  return df
```

```python
# The dataframes
train_df = textfile_to_dataframe(data_path, 'images_family_train.txt', 'train')
valid_df = textfile_to_dataframe(data_path, 'images_family_trainval.txt', 'validation')
test_df = textfile_to_dataframe(data_path, 'images_family_test.txt', 'test')

# Sample of what it looks like 
train_df.head()
```

```python
# Function to crop out copyright information from photo
def crop_photo_tag(path, pix):
    p = Image.open(path)
    w, h = p.size
    left, top, right, bottom = 0, 0, w, h-pix
    crop = p.crop((left, top, right, bottom))
    path_list = path.split('/')
    path_list.insert(-1, 'cropped')
    save_path = ('/').join(path_list)
    crop.save(save_path)
    return
```

```python
# Cropping training pictures 
for idx in range(len(train_df)):
  row = train_df.iloc[idx]
  path = row['path']
  crop_photo_tag(path, 20)
```

```python
# Cropping validation pictures 
for idx in range(len(valid_df)):
  row = valid_df.iloc[idx]
  path = row['path']
  crop_photo_tag(path, 20)
```

```python
# Cropping test pictures 
for idx in range(len(test_df)):
  row = test_df.iloc[idx]
  path = row['path']
  crop_photo_tag(path, 20)
```
